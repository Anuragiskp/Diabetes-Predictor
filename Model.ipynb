```python
import numpy as np 
import pandas as pd
```

```python
dataset = pd.read_csv('./Dataset/fraud test.csv')
dataset.head(5)
```

```python
dataset = dataset.drop(['trans_date_trans_time', 'cc_num', 'trans_num'], axis=1)
dataset.head(5)
```

```python
from datetime import datetime
def calculate_age(dob):
    today = datetime.today()
    dob = datetime.strptime(dob, '%d/%m/%Y')  
    age = today.year - dob.year - ((today.month, today.day) < (dob.month, dob.day))
    return age

dataset['Age'] = dataset['dob'].apply(calculate_age)

dataset= dataset.drop('dob', axis=1)
dataset.head(5)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.compose import ColumnTransformer

columns_to_encode = ['merchant', 'category', 'first', 'last', 'gender', 'street', 'city', 'state', 'job']

label_encoders = {col: LabelEncoder() for col in columns_to_encode}

for col in columns_to_encode:
    dataset[col] = label_encoders[col].fit_transform(dataset[col])
    
dataset.head(5)
```

```python
Y = dataset['is_fraud']
dataset = dataset.drop(['is_fraud'], axis=1)
X = dataset.iloc[:, 1:18 ].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(X_train, Y_train)
```

```python
y_pred = rf_classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(Y_test, y_pred)
print("Accuracy:", accuracy)
```
